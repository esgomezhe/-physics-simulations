```
import numpy as np
import math

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

N = 100
L = 2E-9
sigma = 0.264e-9
r0 = 2**(1/6)*sigma
dt = 1e-5
epsilon = 0.0094/6.242e+18

m = 4e-3/6.023e23
KB = 1.3806488e-23

r = np.zeros((N,3))
for i in range(1,N):
  band0 = 1
  while band0 >= 1:
    band0 = 0
    rt = 2E-9*np.random.rand(1,3)
    for j in range(0, i):
      rij = ((rt[0, 0] - r[j, 0])**2+(rt[0, 1] - r[j, 1])**2+(rt[0, 2] - r[j, 2])**2)**0.5
      if rij <= 2*r0:
        band0 +=1
  r[i] = rt

fig1 = plt.figure()
plt.title('Posiciones iniciales de las particulas')
ax = Axes3D(fig1)
ax.set_zlabel('Z (m)')
ax.scatter3D(r[:,0], r[:,1], r[:,2], 'or')
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.show

v = np.zeros((N,3))
v = 1e-10*np.random.normal(v)
Sum = np.zeros((1,3))
Sum[0,:] = sum(v[:,:])/N
v = v - Sum
Sum[0,:] = sum(v[:,:])/N
v = v - Sum
Sum[0,:] = sum(v[:,:])/N

sumv2 = 0
for i in range(N):
  sumv2 += v[i,0]**2+v[i,1]**2+v[i,2]**2
K = sumv2*m/(2*N)
T = 2*K/(3*KB)
print(N*K,T)
v = (50/T)**(1/2)*v
sumv2 = 0
for i in range(N):
  sumv2 += v[i,0]**2+v[i,1]**2+v[i,2]**2
K = sumv2*m/(2*N)
T = 2*K/(3*KB)
print(N*K,T)

f = np.zeros((N,3))
rij = 0
for i in range(N):
  for j in range(N):
    if i!=j:
      rij = ((r[i,0]-r[j,0])**2+(r[i,1]-r[j,1])**2+(r[i,2]-r[j,2])**2)**(1/2)
      f[i,0] += -24.0*(epsilon*sigma**6*(r[i,0]-r[j,0]))*(1-2*(sigma/rij)**6)/(rij**8)
      f[i,1] += -24.0*(epsilon*sigma**6*(r[i,1]-r[j,1]))*(1-2*(sigma/rij)**6)/(rij**8)
      f[i,2] += -24.0*(epsilon*sigma**6*(r[i,2]-r[j,2]))*(1-2*(sigma/rij)**6)/(rij**8)
a = f/m

for k in range(100):
  ft = f
  for i in range(N):

    r[i,0] += v[i,0]*dt+f[i,0]*dt**2/(2*m)
    r[i,1] += v[i,1]*dt+f[i,1]*dt**2/(2*m)
    r[i,2] += v[i,2]*dt+f[i,2]*dt**2/(2*m)

    rij = 0
    for jj in range(N):
      if i!=jj:
        rij = ((r[i,0]-r[jj,0])**2+(r[i,1]-r[jj,1])**2+(r[i,2]-r[jj,2])**2)**(1/2)
        ft[i,0] += -24.0*(epsilon*sigma**6*(r[i,0]-r[jj,0]))*(1-2*(sigma/rij)**6)/(rij**8)
        ft[i,1] += -24.0*(epsilon*sigma**6*(r[i,1]-r[jj,1]))*(1-2*(sigma/rij)**6)/(rij**8)
        ft[i,2] += -24.0*(epsilon*sigma**6*(r[i,2]-r[jj,2]))*(1-2*(sigma/rij)**6)/(rij**8)

    v[i,0] += (f[i,0]+ft[i,0])*dt/(2*m)
    v[i,1] += (f[i,1]+ft[i,1])*dt/(2*m)
    v[i,2] += (f[i,2]+ft[i,2])*dt/(2*m)

  f = ft

fig2 = plt.figure()
plt.title('Posiciones finales de las particulas')
ax = Axes3D(fig2)
ax.set_zlabel('Z (m)')
ax.scatter3D(r[:,0], r[:,1], r[:,2], 'or')
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.show
```
